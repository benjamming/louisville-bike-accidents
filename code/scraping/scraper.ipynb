```python
import requests
from bs4 import BeautifulSoup as BS

website = "http://crashinformationky.org/Home/WebCrashCodes"
```

```python
req = requests.get(website)
req
```

```python
soup = BS(req.text, "html.parser")
lists = soup.find_all("div", {"class":"col-md-12"})
code_lists = list()
for list in lists:
    title = list.find("div", {"class":"headline"}).h2.text
    content = [line.text for line in list.find_all('li')]
    if content:
        code_lists.append((title, content))
code_lists[3:]
```
